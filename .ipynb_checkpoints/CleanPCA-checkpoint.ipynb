```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from mpl_toolkits import mplot3d
import seaborn as sns
```

```python
# raw data
df = pd.read_csv('Fires.csv')
```

```python
test = df[['DISCOVERY_DOY',
        'DISCOVERY_TIME',
        'STAT_CAUSE_CODE',
        'STAT_CAUSE_DESCR',
        'FIRE_SIZE',
        'LATITUDE',
        'LONGITUDE',
        'STATE']]

# convert state to numerical value
le = preprocessing.LabelEncoder()
test['STATE'] = le.fit_transform(test['STATE'])

# drop rows with at least one missing value
test = test.dropna()
```

```python
# PCA PREP!!

features = ['DISCOVERY_DOY',
            'DISCOVERY_TIME',
            #'STAT_CAUSE_CODE', this is our y
            'FIRE_SIZE',
            'LATITUDE',
            'LONGITUDE',
            'STATE']

# Separating out the features
x = test.loc[:, features].values

# Standardizing the features
x = StandardScaler().fit_transform(x)
```

```python
# PCA with 2 Components

pca2 = PCA(n_components=2)
principalComponents2 = pca2.fit_transform(x)

principalDf2 = pd.DataFrame(data = principalComponents2
             , columns = ['PCA0', 'PCA1'])
```

```python
# add the y/target/class identifier to the dataframe (for ~color~!)
finalDf2 = pd.concat([principalDf2, test['STAT_CAUSE_DESCR']], axis = 1)
finalDf2 = finalDf2.dropna()

# make it categorical data so hue will accept it :3
finalDf2['STAT_CAUSE_DESCR'] = pd.Categorical(finalDf2.STAT_CAUSE_DESCR)
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x=finalDf2['PCA0'], y=finalDf2['PCA1'],hue=finalDf2['STAT_CAUSE_DESCR'].tolist(), alpha=0.5)

plt.title("2 component PCA")
```

```python
# PCA with 3 components

pca1 = PCA(n_components=3)
principalComponents1 = pca1.fit_transform(x)
principalDf1 = pd.DataFrame(data = principalComponents1
             , columns = ['PCA0', 'PCA1', 'PCA2'])
```

```python
test['STAT_CAUSE_CODE']=pd.Categorical(test['STAT_CAUSE_CODE'])
my_color=test['STAT_CAUSE_CODE'].cat.codes
```

```python
ex_variance=np.var(principalComponents1,axis=0)
ex_variance_ratio = ex_variance/np.sum(ex_variance)
ex_variance_ratio
```

```python
# Plot initialisation for 3 component PCA

fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111, projection='3d')

ax.set_xlabel('PC 1', fontsize = 15)
ax.set_ylabel('PC 2', fontsize = 15)
ax.set_zlabel('PC 3', fontsize = 15)
ax.set_title('3 component PCA', fontsize = 20)

# default angle view is (-60,3)
ax.scatter(principalDf1['PCA0'], principalDf1['PCA1'], principalDf1['PCA2'], c=my_color, cmap="Set2_r", s=30, alpha = 0.7)

# can't figure out how to put on a legend sad
ax.grid()
```

```python
# Plot initialisation for 3 component PCA

fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111, projection='3d')

ax.set_xlabel('PC 1', fontsize = 15)
ax.set_ylabel('PC 2', fontsize = 15)
ax.set_zlabel('PC 3', fontsize = 15)
ax.set_title('3 component PCA', fontsize = 20)

# some anglez
ax.view_init(30, -30)

ax.scatter(principalDf1['PCA0'], principalDf1['PCA1'], principalDf1['PCA2'], c=my_color, cmap="Set2_r", s=30, alpha = 0.7)

ax.grid()
```

```python
# Plot initialisation for 3 component PCA

fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111, projection='3d')

ax.set_xlabel('PC 1', fontsize = 15)
ax.set_ylabel('PC 2', fontsize = 15)
ax.set_zlabel('PC 3', fontsize = 15)
ax.set_title('3 component PCA', fontsize = 20)

# another angle
ax.view_init(0, -90)

ax.scatter(principalDf1['PCA0'], principalDf1['PCA1'], principalDf1['PCA2'], c=my_color, cmap="Set2_r", s=30, alpha = 0.7)

ax.grid()
```
